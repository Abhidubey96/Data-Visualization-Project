### I am directly reading website and fetching out tables with pandas

```python
import pandas as pd

championsleaguefinal_url = 'https://www.stadiumguide.com/figures-and-statistics/lists/champions-league-final-venues/'

dfs = pd.read_html(championsleaguefinal_url)

championsleaguefinalstadium_df = dfs[0]
```

```python
championsleaguefinalstadium_df
```

#### Some cleaning required for this data set

```python
championsleaguefinalstadium_df=championsleaguefinalstadium_df.sort_values(by=['Year'])

championsleaguefinalstadium_df=championsleaguefinalstadium_df.reset_index(drop=True)

championsleaguefinalstadium_df
```

### We have few questions regarding our data set

#### Like which match is most popular?
#### How much is the actual strength ( capacity ) of stadium and how many attendees we have for that match?

##### Assumption : More number of attendees watching match means match is more popular or team is more popular

#### So with the help of this we will analyse which team is most popular in champions league tournament

### To find above answers i am using another data set which will help me to analyse the strength capacity of stadium.

### This Data Source is also public
#### https://www.stadiumguide.com/figures-and-statistics/lists/europes-largest-football-stadiums/

```python
import pandas as pd

stadium_url = 'https://www.stadiumguide.com/figures-and-statistics/lists/europes-largest-football-stadiums/'

dfs = pd.read_html(stadium_url)

df_stadium= dfs[0]
```

```python
df_stadium
```

#### Some cleaning required for this data set

### Now i am merging two data frames to collaborate the analysis

```python
df_stadiumcapacity=championsleaguefinalstadium_df.merge(df_stadium, on='Stadium', how='left')

df_stadiumcapacity
```

#### Some cleaning required for this data set

```python
df_stadiumcapacity.drop(columns=['Unnamed: 0', 'Match','City','Country'], inplace=True)
```

```python
df_stadiumcapacity.tail(7)
```

```python
df_stadiumcapacity=df_stadiumcapacity.drop(df_stadiumcapacity.tail(6).index)

df_stadiumcapacity
```

### Now i am adding all data frame together 

### Final Data Frame is df_final

```python
df_final = pd.concat([df_championsleague, df_stadiumcapacity], axis=1, sort=False)

df_final
```

#### Some cleaning required for this data set


```python
df_final.drop(columns=['Season', 'Venue'], inplace=True)

df_final
```

#### Note: Some of the stadiums are not is use or they are out of service so because of that for some scenarios i am using attendance = capacity



```python
df_final['Capacity']= df_final['Capacity'].fillna(df_final['Attendance'])
```

### Lets Convert it to int so that there will be no float values

```python
df_final['Capacity']=df_final['Capacity'].astype(int)
```

```python
df_final
```

### Our Final Cleaned Data Frame is Ready for visualization

### Data Exploration

```python
df_visual = df_final
```

#### Creating total goals column by adding winner score and finalist score

```python
df_visual["Match Goals"] = df_final["Winner Score"] + df_final["Finalist Score"]
```

```python
df_visual
```

```python
df_visual["Match Played"] = df_final["Winner"] + " vs " + df_final["Finalist"]
```

```python
df_visual.head(35)
```

## Visualize Data and analyse Most Scoring Match in the history of Champions Leauge ?



```python
plt.figure();
plt.figure(figsize=(30,15))

plt.bar(df_visual["Match Played"], df_visual["Match Goals"])
plt.title('Most Scored Match in Champions League History')
plt.xlabel('Match Played')
plt.ylabel('Total Goals')
plt.xticks(rotation=90)

plt.show()
```

### Most scoring match played between Real Madrid vs Eintracht Frankfurt with 10 Goals Total

```python
df_team=df_visual.groupby("Winner")["Winner Score"].sum() 

df_team
```

## Visualize Data and Analyse Most Scoring Team in the history of Champions Leauge ?

```python
df_team.plot(kind="bar", figsize=(15,8))
```

### Real Madrid Scores the most number of goals total 35

```python
df_stadium = df_visual 
df_stadium["people attendent match in excess"] = df_visual["Attendance"] - df_visual["Capacity"]
```

```python
df_stadium
```

```python
df_stadium=df_stadium[["Winner","Year","people attendent match in excess"]]
```

```python
df_stadium["Team with year"]= df_stadium["Winner"] + " : " + df_stadium["Year"].map(str)
```

```python
df_stadium.drop(['Winner', 'Year'], axis=1, inplace = True)
```

## Visualize Data and Analyse when stadium is most occupied by people ?
